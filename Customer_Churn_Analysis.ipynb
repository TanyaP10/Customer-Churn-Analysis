```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create cross-tab of counts
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentages row-wise
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Get pastel colors from seaborn
pastel_colors = sns.color_palette("pastel")

# Plot stacked percentage bar chart with pastel colors
ax = ct_pct.plot(kind='bar', stacked=True, figsize=(4,4), color=pastel_colors)

plt.title("Churn By Senior Citizen (in %)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor=(0.9,0.9))

# Add percentages on each bar section
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.show()
```

comparitively a greater percentage of senior citizen category have churned out

```python

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(
        x=col,
        data=df,
        ax=axes[i],
        hue=df["Churn"],
        palette="pastel"  
    )
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#Most customers use PhoneService and a large share have MultipleLines, with churn being higher among those with additional services. Fiber optic internet users show higher churn compared to DSL and no-internet groups. Customers without OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport are more likely to churn than those with these services. Streaming services (TV/Movies) also show a higher churn rate when subscribed, indicating add-on services may correlate with customer dissatisfaction.
